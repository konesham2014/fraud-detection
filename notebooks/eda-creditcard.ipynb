```python
plt.figure(figsize=(7,5))

sns.countplot(
    x="Class",
    data=credit
)

plt.title(
    "Credit Card Fraud Class Distribution"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    credit["Amount"],
    bins=50,
    kde=True
)

plt.title(
    "Transaction Amount Distribution"
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=credit["Amount"]
)

plt.title(
    "Transaction Amount Outliers"
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=credit["Amount"]
)

plt.title(
    "Transaction Amount Outliers"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    credit["Time"],
    bins=50,
    kde=True
)

plt.title(
    "Transaction Time Distribution"
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=credit["Time"]
)

plt.title(
    "Transaction Time Outliers"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Class",
    y="Amount",
    data=credit
)

plt.title(
    "Fraud vs Transaction Amount"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Class",
    y="Time",
    data=credit
)

plt.title(
    "Fraud vs Transaction Time"
)

plt.show()
```

```python
corr = credit.corr()

plt.figure(figsize=(18,12))

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0
)

plt.title(
    "Credit Card Dataset Correlation Matrix"
)

plt.show()
```

```python
fraud_corr = (

    credit.corr()["Class"]

    .sort_values(
        ascending=False
    )

)

fraud_corr
```

```python
fraud_corr.head(10)
```

```python
fraud_corr.tail(10)
```

```python
plt.figure(figsize=(10,8))

fraud_corr.drop("Class").plot(
    kind="barh"
)

plt.title(
    "Feature Correlation with Fraud"
)

plt.show()
```
